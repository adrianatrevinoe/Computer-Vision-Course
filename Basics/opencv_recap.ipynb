```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('DATA/00-puppy.jpg') # converts to numpy array
type(img)
plt.imshow(img)
```

```python
img.shape
```

MATPLOTLIB --> expects RGB

OPENCV --> expects BGR

```python
bgr_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(bgr_img)
```

```python
gray_img = cv2.imread('DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img,cmap='gray')
```

```python
bgr_img.shape
# height, width, bands
```

```python
new_img = cv2.resize(bgr_img,(1100,800)) # (width,height)
print(new_img.shape)
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.6
new_img = cv2.resize(new_img,(0,0),new_img,w_ratio,h_ratio)
print(new_img.shape)
plt.imshow(new_img)
```

```python
flipped = cv2.flip(bgr_img,-1) # 0 vertical, 1 horizontal, -1 both
plt.imshow(flipped)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(340,10),pt2=(500,150),color=(0,255,0),
              thickness=(10))
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(255,0,255),
              thickness=(10))
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),
           thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(img=blank_img,pt1=(0,0),pt2=(512,512),color=(102,255,255),
         thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,400),fontFace=font,
            fontScale=5,color=(255,255,255),thickness=3,
            lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300]],dtype=np.int32)
vertices.shape
```

```python
pts = vertices.reshape(-1,1,2)
pts.shape
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,255),
              thickness=5)
plt.imshow(blank_img)
```
