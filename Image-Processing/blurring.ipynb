# Blurring
Gets rid of noise

Smoothing

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('../DATA/bricks.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
display_img(img)
```

## Gamma Correction

Changes brightness

```python
# val < 1 : brighter
# 1 > : darker
gamma = 1/4

result = np.power(img,gamma)
display_img(result)
```
